# Implementation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
print(iris['target'])
```

```python
print(iris['data'])
```

```python
X = pd.DataFrame(iris['data'],columns=iris['feature_names'])
```

```python
X.head()
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train,y_train)
```

# Visualize the Decision Tree

```python
from sklearn import tree
plt.figure(figsize=(25,8))
tree.plot_tree(treeclassifier,filled=True)
plt.show()
```

# Accuracy of the Model

```python
y_pred = treeclassifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

# Model at maxdepth = 2

```python
treeclassifier2 = DecisionTreeClassifier(max_depth=2)
```

```python
treeclassifier2.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(25,8))
tree.plot_tree(treeclassifier2,filled=True)
plt.show()
```

```python
y_pred2 = treeclassifier2.predict(X_test)
```

```python
print(confusion_matrix(y_pred2,y_test))
print(accuracy_score(y_pred2,y_test))
print(classification_report(y_pred2,y_test))
```

# Decision Tree Classifier Prepruning and Hyperparameter Tuning

```python
param = {
    'criterion':['gini', 'entropy','log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto','sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treemodel,param_grid=param,cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
